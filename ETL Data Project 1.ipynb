```python
#Computing ID: om4kud
# libraries
import json
import requests
import sqlite3
import pandas as pd
```

```python
#Ingest local csv file
adidas_usa = pd.read_csv("adidas_usa.csv")
```

```python
#connect to the database
connection = sqlite3.connect("Adidas Database.db") 
```

```python
#Check to see what the original csv file looks like
display(adidas_usa)
```

```python
output = input("Please enter a data output format: ")
```

```python
#Error handling for valid inputs
try:
    if output == "sql":
        adidas_usa.to_sql('adidas_usa', connection)
    if output == "json":
        adidas_usa.to_json (r'adidas_usa.json')
    else:
        print("Error: Please enter a valid data output format, either json or sql, and try again.")
except ValueError:
    print("This database table has already been created.")
    
```

```python
#read the SQL Query from database into a pandas dataframe
sql_string = 'SELECT * FROM adidas_usa'
adidas_usa = pd.read_sql(sql_string, connection)
```

```python
#check to make sure that it worked
display(adidas_usa)
```

```python
#Modify number of columns: Transformation Aspect of ETL
#Multiply average_rating by reviews_count to get Rating Reviews
adidas_usa['Rating Reviews'] = adidas_usa['average_rating'] * adidas_usa['reviews_count']
```
